```python
import pandas as pd
import datetime as dt
```

```python
ds_salaries_csv = "ds_jobs.csv"
```

```python
ds_salaries_df = pd.read_csv(ds_salaries_csv)
```

```python
ds_salaries_df.head()
```

```python
ds_salaries_df.count()
```

```python
ds_salaries_df.isnull().sum()
```

```python
ds_salaries_df.tag.unique()
```

```python
ds_salaries_df = ds_salaries_df.rename(columns={'tag':'area'})
```

```python
ds_salaries_df.head()
```

```python
ds_salaries_df.Education.unique()
```

```python
# Education column is redundant. Drop

ds_salaries_df.drop(['Education'], axis = 1, inplace = True, errors = 'ignore')
```

```python
ds_salaries_df.head()
```

```python
ds_salaries_df.State.unique()
```

```python
# Rename Location as City
ds_salaries_df = ds_salaries_df.rename(columns={'location':'City'})
```

```python
# 'Gender' columns has 68% of missing values. Drop the column. 

ds_salaries_df.drop(['gender'], axis = 1, inplace = True, errors = 'ignore')

# 'cityid' column is not relevant 

ds_salaries_df.drop(['cityid'], axis = 1, inplace = True, errors = 'ignore')

# 'rowNumber' column is not relevant

ds_salaries_df.drop(['rowNumber'], axis = 1, inplace = True, errors = 'ignore')

# 'dmaid' column is not relevant

ds_salaries_df.drop(['dmaid'], axis = 1, inplace = True, errors = 'ignore')

# drop all columns related to race, as it is not significant for our analysis

ds_salaries_df.drop(['Race_Asian'], axis = 1, inplace = True, errors = 'ignore')
ds_salaries_df.drop(['Race_White'], axis = 1, inplace = True, errors = 'ignore')
ds_salaries_df.drop(['Race_Two_Or_More'], axis = 1, inplace = True, errors = 'ignore')
ds_salaries_df.drop(['Race_Black'], axis = 1, inplace = True, errors = 'ignore')
ds_salaries_df.drop(['Race_Hispanic'], axis = 1, inplace = True, errors = 'ignore')
ds_salaries_df.drop(['Race'], axis = 1, inplace = True, errors = 'ignore')
```

```python
ds_salaries_df.head()
```

```python
# Rename Salary Level as Salary_lvl
ds_salaries_df = ds_salaries_df.rename(columns={'Salary Level':'Salary_lvl'})
```

```python
ds_salaries_df.head()
```

```python
print(ds_salaries_df['level'].value_counts())
```

```python
# Rename Salary Level as Salary_lvl
ds_salaries_df = ds_salaries_df.rename(columns={'Salary_lvl':'Salary Level'})
```

```python
# Drop 'level' column. The information is not consistent 

ds_salaries_df.drop(['level'], axis = 1, inplace = True, errors = 'ignore')
```

```python
ds_salaries_df.head()
```

```python
# Rename Salary_lvl as Salary Level
ds_salaries_df = ds_salaries_df.rename(columns={'Salary_lvl':'Salary Level'})
```

```python
ds_salaries_df.isnull().sum()
```

```python
# Delete the rows with missing values in the 'company' column 

ds_salaries_df = ds_salaries_df[ds_salaries_df['company'].notna()]
```

```python
ds_salaries_df.isnull().sum()
```

```python
# Keep only the year from the 'timestamp' column

ds_salaries_df['timestamp'] = pd.DatetimeIndex(ds_salaries_df['timestamp']).year
```

```python
# Check values in the 'timestamp' column

ds_salaries_df.timestamp.unique()
```

```python
# Change the name of the column, from 'timestamp' to 'year'

ds_salaries_df = ds_salaries_df.rename(columns={'timestamp':'Year'})
```

```python
ds_salaries_df.head()
```

```python
# Rearrange the order of the columns

neworder = ['Year','company','City','State','title','Salary Level', 'area', 'totalyearlycompensation', 'yearsofexperience', 'yearsatcompany', 'basesalary', 'stockgrantvalue', 'bonus', 'Masters_Degree', 'Bachelors_Degree', 'Doctorate_Degree', 'Highschool', 'Some_College']
ds_salaries_df=ds_salaries_df.reindex(columns=neworder)
ds_salaries_df.head()
```

```python
# Rename the names of the columns. Add capital letters

ds_salaries_df = ds_salaries_df.rename(columns={'company':'Company','title':'Title','totalyearlycompensation':'Total Annual Compensation','yearsofexperience':'Years of Experience','yearsatcompany':'Years at Company','basesalary':'Base Salary','stockgrantvalue':'Stock Grant Value','bonus':'Bonus','Masters_Degree':'Masters Degree','Bachelors_Degree':'Bachelors Degree','Doctorate_Degree':'Doctorate Degree','Some_College':'Some College', 'area':'Area'})

ds_salaries_df.head()
```

```python
import os   
ds_salaries_df.to_csv('ds_salaries_clean.csv') 
```
